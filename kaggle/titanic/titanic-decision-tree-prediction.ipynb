```python
import pandas as pd
import numpy as np
 
train = pd.read_csv("./data/train.csv")
test = pd.read_csv("./data/test.csv")
```

```python
def kesson_table(df): 
        null_val = df.isnull().sum()
        percent = 100 * df.isnull().sum()/len(df)
        kesson_table = pd.concat([null_val, percent], axis=1)
        kesson_table_ren_columns = kesson_table.rename(
        columns = {0 : '欠損数', 1 : '%'})
        return kesson_table_ren_columns
 
kesson_table(train)
kesson_table(test)
```

```python
# 1. 欠損データに代理データを埋める（様々な方法があるが、今回は中央値を使う）
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")

kesson_table(train)
```

```python
# 2. カテゴリカル・データを数字化
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

train.head(10)
```

```python
# 3. テストデータも同様にする
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].median())

test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2

test.head(10)
```

```python
# 予想モデル: 1 決定木
# 「チケットクラス・性別・年齢・料金」 から予想してみる
from sklearn import tree

# train: 目的変数 (正解データ) と説明変数 (教師データ)の値を取得
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values

# 決定木
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)

# test: 説明変数の値を取得
test_features = test[["Pclass", "Sex", "Age", "Fare"]].values

# test: 予想
my_prediction = my_tree_one.predict(test_features)
```

```python
print(my_prediction)
```

```python
# CSV出力
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"])
my_solution.to_csv("my_tree_one.csv", index_label = ["PassengerId"])
```

```python
# 予想モデル: 2 決定木
# 「チケットクラス・性別・年齢・料金・配偶者/兄弟の数・親/子供の数・出発港」 から予想してみる

# train: 説明変数
features_two = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values

# 決定木
my_tree_two = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 5, random_state = 1)
my_tree_two = my_tree_two.fit(features_two, target)

# test: 説明変数
test_features_two = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values

# test: 予想
my_prediction_two = my_tree_two.predict(test_features_two)
```

```python
# CSV出力
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution_two = pd.DataFrame(my_prediction_two, PassengerId, columns = ["Survived"])
my_solution_two.to_csv("my_tree_two.csv", index_label = ["PassengerId"])
```

```python
# 作成した決定木を可視化 (pydotplus パッケージを利用)
import pydotplus
from sklearn.externals.six import StringIO
dot_data = StringIO()
tree.export_graphviz(my_tree_one, out_file=dot_data)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
 
# PDFファイルに出力
graph.write_pdf("graph.pdf")
```

```python
# 予想モデル: 3 決定木
# 「チケットクラス・性別・年齢・料金・配偶者/兄弟の数・親/子供の数・客室番号・出発港」 から予想してみる
# NOTE: 客室番号は pd.factorizeで同じ文字列同士でまとあげ、数値列化する。
train["Cabin"], _ = pd.factorize(train["Cabin"])
test["Cabin"], _ = pd.factorize(test["Cabin"])

# train: 説明変数
features_three = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Cabin", "Embarked"]].values

# 決定木
my_tree_three = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 5, random_state = 1)
my_tree_three = my_tree_three.fit(features_three, target)

# test: 説明変数
test_features_three = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Cabin", "Embarked"]].values

# test: 予想
my_prediction_three = my_tree_three.predict(test_features_three)
```

```python
# CSV出力
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution_three = pd.DataFrame(my_prediction_three, PassengerId, columns = ["Survived"])
my_solution_three.to_csv("my_tree_three.csv", index_label = ["PassengerId"])
```
